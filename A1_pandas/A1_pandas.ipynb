```python
import pandas as pd
import numpy as np
PATH = 'athlete_events.csv'
df = pd.read_csv(PATH)
df.head()
```

```python
data =  df[df['Games'] == '1992 Summer'][['Sex','Age','Games']]
male = int(data[data['Sex'] == 'M'].min()['Age'])
female = int(data[data['Sex'] == 'F'].min()['Age'])
print(str(male) + ' and ' + str(female))
```

```python
male_participants = df[(df['Sex'] == 'M') & (df['Games'] == '2012 Summer')].drop_duplicates(subset = 'ID')
male_basketball_players = male_participants[male_participants['Sport'] == 'Basketball']
round(male_basketball_players.size/male_participants.size * 100, 1)
```

```python
data = df
data['Age15'] = (15 <= df['Age']) & (df['Age'] < 25)
data['Age25'] = (25 <= df['Age']) & (df['Age'] < 35)
data['Age35'] = (35 <= df['Age']) & (df['Age'] < 45)
data['Age45'] = (45 <= df['Age']) & (df['Age'] <= 55)
data_map = dict([
(data[data['Age15'] & (data['Year'] == 2008)]['Age15'].size,"[15-25]"),
(data[data['Age25'] & (data['Year'] == 2008)]['Age25'].size,"[25-35]"),
(data[data['Age35'] & (data['Year'] == 2008)]['Age35'].size,"[35-45]"),
(data[data['Age45'] & (data['Year'] == 2008)]['Age45'].size,"[45-55]")])

"" + data_map[min(data_map.keys())] + " and " + data_map[max(data_map.keys())]
```

```python
summer = (df[(df['Season'] == 'Summer') & (df['City'] == 'Atlanta')].size > 0)
winter = (df[(df['Season'] == 'Winter') & (df['City'] == 'Squaw Valley')].size > 0)

str(summer) + "," + str(winter)
```
